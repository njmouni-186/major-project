## Displaying the Courses Data Set

```python

std=pd.read_csv('students.csv')

x=std.drop(["Serial No.","Student ID","PASS/FAIL"],axis=1)
y=std["PASS/FAIL"]
x.head()
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)  
pclassifier = LogisticRegression(random_state = 0)
pclassifier.fit(X_train, y_train)
y_pred=pclassifier.predict(X_test)
x.head()
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.model_selection import cross_val_score

print('Logistic Regression Classifier')
print('accuracy :',accuracy_score(y_test,y_pred))
print('F1 score :',f1_score(y_test,y_pred))
print('Precision:',precision_score(y_test,y_pred))
print('Recall   :',recall_score(y_test,y_pred))
print('Cross validation score : \n',cross_val_score(estimator = pclassifier, X = X_train, y = y_train, cv =5))
print('Mean of cross validation score : ',cross_val_score(estimator = pclassifier, X = X_train, y = y_train, cv =5).mean())
```
